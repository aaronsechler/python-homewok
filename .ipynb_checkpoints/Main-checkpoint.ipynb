```python
#import pathlab and csv library
from pathlib import Path
import csv

#set the file path. I currently have a problem with Github and or jupyterlabs that prevents me from getting the file that ...
# way so I have downloaded it and manually entered the path.

csvpath = Path("/Users/aaronbsechler/Desktop/02-Homework_02-Python_Instructions_PyBank_Resources_budget_data.csv")

#Initialize the list of records

total_months = 0
total_profit = 0
months_with_a_profit = 0
months_with_a_loss = 0
months_with_no_change = 0
greatest_profit = 0
month_with_greatest_profit = None
greatest_loss = 0
month_with_greatest_loss = None
previous_change_in_profitablity = 0
change_in_profitability = 0
sum_of_changes_in_profitability = 0
average_change_in_profitability = 0

# open the csv file as an object
# pass in the csv files to the csv.reader(function) with a "," as the delimitor and return the csv reader object. used next because this file has a header.
# the next set of code removes the header
# the for loop needs two pieces of data to compute the change in profitability, so the first_row=next removes that from the loop.
# the two next commands remove the header and the first set of values from the for loop, and the two lines of codes after them includes that data in the total months, the total profit, and the total months with profit or loss.
# the following code makes sure that the first row of data is included in the months and total profit
# setting greatest profit and greatest loss to the first row makes sure that value is compared to the other values, this way is shorter than us if then logic.
# the if then logic includes the first row of data in the logic that counts the number of profitable or unprofitable months
# create a for loop that compiles the total number of months, the number of profitable or unprofitable months,...
# sets minimum and maximums for the change, and creates a new list of changes in profitability

with open(csvpath, 'r') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    csv_header = next(csvreader)
    first_row = next(csvreader)
    total_months += 1
    total_profit = total_profit + float(first_row[1])
    previous_change_in_profitability = float(first_row[1])
    greatest_profit = float(first_row[1])
    greatest_loss = float(first_row[1])
    if float(first_row[1]) > 0:
        months_with_a_profit += 1
    elif float(first_row[1]) < 0:
        months_with_a_loss += 1
    else:
        months_with_no_change +=1
    for row in csvreader:
        date = row[0]
        profit = float(row[1])
        total_months += 1
        total_profit = total_profit + profit
        if profit > 0:
            months_with_a_profit = months_with_a_profit + 1
        elif profit < 0:
            months_with_a_loss = months_with_a_loss + 1
        elif profit == 0: 
            months_with_no_change + 1
        if profit > greatest_profit:
            greatest_profit = profit
            month_with_greatest_profit = date
        elif profit < greatest_loss:
            greatest_loss = profit
            month_with_greatest_loss = date 
        change_in_profitability = profit - previous_change_in_profitablity
        sum_of_changes_in_profitability = sum_of_changes_in_profitability + change_in_profitability
        previous_change_in_profitability = change_in_profitability
        
      

    
#takes the average of the sum of changes in profitability

average_change_in_profitability = sum_of_changes_in_profitability / (total_months-1)

# print results to the terminal 

print("Summary of Results")
print(f"The file has {total_months} months in it. There were {months_with_a_profit} months with a profit, and {months_with_a_loss} months with a loss.")
if months_with_no_change > 0:
    print(f"There were {months_with_no_change} months where the operation broke even.")
print(f"The total amount of profit was ${total_profit:,.2f}.")
print(f"The greatest monthly profit was ${greatest_profit:,.2f} in {month_with_greatest_profit}, and the greatest monthly loss was ${greatest_loss:,.2f} in {month_with_greatest_loss}.")
print(f"The average change in profitability was ${average_change_in_profitability:,.2f}.")


# print the results to a text file

with open ("results.txt", "w+") as textfile:
    textfile.write("Summary of Results")
    textfile.write(f"The file has {total_months} months in it. There were {months_with_a_profit} months with a profit, and {months_with_a_loss} months with a loss.")
    if months_with_no_change > 0:
        textfile.write(f"There were {months_with_no_change} months where the operation broke even.")
    textfile.write(f"The total amount of profit was ${total_profit:,.2f}.")
    textfile.write(f"The greatest monthly profit was ${greatest_profit:,.2f} in {month_with_greatest_profit}, and the greatest monthly loss was ${greatest_loss:,.2f} in {month_with_greatest_loss}.")
    textfile.write(f"The average change in profitability was ${average_change_in_profitability:,.2f}.")
f.close()


```
